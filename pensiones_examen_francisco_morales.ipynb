```python
#Tasas
inte=float(input("¿Cuál es la tasa de interés (en decimal) (i)? : "))
j=float(input("¿Cuál es la tasa de incremento salarial (en decimal) (j)? : "))
inf=float(input("¿Cuál es la tasa de inflación (en decimal) (inf)? : "))
#Anualidad
Dr=float(input("¿Cuál es el valor de Dr? : "))
Nr=float(input("¿Cuál es el valor de Nr? : "))
```

Anualidad anual
$$
\ddot{a}_{r}= \dfrac{N_r}{D_r}
$$
Anualidad mensual
$$
\ddot{a}_{r}^{(12)}= \dfrac{N_r}{D_r}-\dfrac{11}{24}
$$

Woolhouse

$$
\ddot{a}_{r}^{(m)}= \dfrac{N_r}{D_r}-\dfrac{2m-1}{2m}
$$

```python
#Anualidad
ax=int(input("¿Cómo es tu anualidad (0 Anual y 1 Mensual)? : "))
if ax==0:
    anua=Nr/Dr
    print(f"La anualidad vale {anua}")

else :
    anua=Nr/Dr-11/24
    print(f"La anualidad vale {anua}")
```

```python
import string
# Listas vacia
empleado = []
salarios = []
y= []
x= []
r= []

#Número de empleados
num_emp = int(input("¿Cuántas personas quieres pensionar? "))
#Datos de cada empleado
for i in range(num_emp):
    print(f"\n Información del empleado {string.ascii_uppercase[i]}:")
    edad_inicial= int(input("¿Cuál es la edad inicial? "))
    edad_actual= int(input("¿Cuál es la edad actual? "))
    edad_retiro= int(input("¿Cuál es la edad de retiro? "))
    salario = int(input("¿Cual es su salario actual? "))

    # Guardar datos en listas
    y.append(edad_inicial)
    x.append(edad_actual)
    r.append(edad_retiro)
    salarios.append(salario)

columnas=['y','x','r','Salario_x']

#Indice con letas A, B , C ....
filas=list(string.ascii_uppercase[:num_emp])

# Crear DataFrame
df = pd.DataFrame({
    "y": y,
    "x": x,
    "r": r,
    "Salario_x": salarios,
})

# Mostrar DataFrame
df.index=filas
```

```python
#Ajuste si es anual o mensual
pago=int(input("¿Cóomo ingresaste el salario (0 Mensual y 1 Anual)? : "))

if pago==0:
    a=12*anua
else:
    a=anua
```

# **Z%**📋
---

**Nota:** $z%$ cambia de acuerdo al beneficio que se plantea en el problema por lo que $z%$ sera diferente para cada plan de pensiones.

En este caso el beneficio es: $6$ meses de sueldo más $1$ mes de sueldo por cada año de servicio

Beneficio que premia antiguedad $ z  \% $
$$
z \ \% \  12\ddot{a}_{r} (r-y) Sr \ = \ [6+1(r-y)]Sr
$$

$$
z  \ \%  \ = \ \dfrac{ [6+1(r-y)]} {12\ddot{a}_{r} (r-y)}
$$

```python
#Calculo de z%
df['z%']=round((6+df['r']-df['y'])/(12*a*(df['r']-df['y'])),6)
#Redondeo a 6 decimales
z=round(float(df['z%'].mean()),6)
z
```

```python
print(f"El valor de z es {z*100} % ")
```

```python
print(f"El beneficio es calculado como {z*100} % * (r-y) * Sr")
```

# **Valuación actuarial**🧠
---

Salario retiro

$$
S_r=S_x*(1+j)^{(r-x)}
$$

```python
#Calculo de x-y,r-y , x-y y Sr
df['x-y']=df['x']-df['y']
df['r-x']=df['r']-df['x']
df['r-y']=df['r']-df['y']
df['Sr']=round(df['Salario_x']*(1+j)**df['r-x'],6)
df
```

Beneficio pensión
$$
B_r=(r-y)*z\% S_r
$$

Beneficio actual
$$
B_x=(x-y)*z\% S_r
$$

Beneficio anual
$$
b_t=z\% S_r
$$

Factor de descuento
$$
v^{(r-x)}=\left( \dfrac{1}{1+i}\right)^{(r-x)}
$$

Probabilidad de llegar a edad de retiro
$$
_{r-x}p_{x}^{(τ)}=\dfrac{l_{r}^{(τ)}}{l_{x}^{(τ)}}
$$

```python
#Calculo de Br,Bx y bt
df['Br']=z*df['Sr']*df['r-y']
df['Bx']=z*df['Sr']*df['x-y']
df['bt']=z*df['Sr']
df
```

```python
#Calculo de v^(r-x)
df['v^(r-x)']=(1/(1+inte))**df['r-x']
df
```

```python
#Calculamos los decrementos múltiples en caso de existir
de=int(input("¿Existen decremento antes del retiro (0 No y 1 Si)? : "))
if de==0:
    #Columnas de 1's
    df['r-xPx (τ)']=1
else:
    #Calculo con l_x's
    lr=float(input("¿Cuál es el valor de l_65? : "))
    #Vector vacio
    probas= [ ]
    for i in range(num_emp):
      li=float(input(f"¿Cuál es el valor de l_{df['x'].iloc[i]}?:"))
      p_i=lr/li
      #Se agrega cada probabilidad al vector
      probas.append(p_i)
    #Se añade al dataframe
    df['r-xPx (τ)']=probas

df['r-xPx (τ)']
```


$$
PVB_r= B_r*\ddot{a}_{r}
$$


$$
PVB_x= B_r*\ddot{a}_{r}*v^{(r-x)}*_{r-x}p_{x}^{(τ)}
$$

Valor presente del benifico acumulado
$$
AL_x= B_x*\ddot{a}_{r}*v^{(r-x)}*_{r-x}p_{x}^{(τ)}
$$

Valor presente de un año definido
$$
NC_x= b_t*\ddot{a}_{r}*v^{(r-x)}*_{r-x}p_{x}^{(τ)}
$$

```python
#Calculamos PVFBr, PVBx, ALx y NCx
df['PVFBr']=a*df['Br']
df['PVFBx']=a*df['Br']*df['v^(r-x)']*df['r-xPx (τ)']
df['ALx']=a*df['Bx']*df['v^(r-x)']*df['r-xPx (τ)']
df['NCx']=a*df['bt']*df['v^(r-x)']*df['r-xPx (τ)']
df
```

```python
# Calculamos la suma
suma = df.sum()
# Agregamos la fila de suma
df.loc['Total'] = suma
df
```

#**Contabilidad** 🗃️

---

```python
#Datos para la contabilidad: ALx, NCx, Fx, años de amotización
#desviaciones y modificaciones
ALx=df['ALx']['Total']
NCx=df['NCx']['Total']
Fx=float(input("¿Cuál es el valor de Fx? : "))
años=int(input("¿En cuántos años vas a amortizar? : "))
des=float(input("¿Cuál es el valor de las desviaciones? : "))
mod=float(input("¿Cuál es el valor de las modificaciones? : "))
res=Fx+(Fx-ALx)+ALx-Fx+des+mod
```

```python
#Desfinir dataframe
filas=['Contabilidad','ALx','Fx','Fx-ALx','ALx-Fx','Desviaciones',
       'Modificaciones','Reservas']
columnas=['Respuesta']
#Mostrar resultados calculados
respuestas=['Situación financiera real en x', ALx, Fx,
            Fx-ALx, ALx-Fx, des, mod, res
            ]
df_balance=pd.DataFrame(data=respuestas, columns=columnas, index=filas)
print("El balance actuarial es:")
#Mostrar respuestas
df_balance
```

```python
#Datos para el costo de la contabilidad
#ALx,NCx, costo
ALx=df['ALx']['Total']
NCx=df['NCx']['Total']
costo=NCx*(1+inte)+inte*ALx-inte*Fx+ALx/años+des/años+mod/años
#Definir dataframe
filas=['NCx*(1+i)','i*ALx','-i*Fx','Amortizar',f'UALx/{años}',f'Desv/{años}',
       f'Mod/{años}','Costo anual']
columnas=['Respuesta']
#Asignar valores calculados
respuestas=[NCx*(1+inte), inte*ALx, -inte*Fx,
            f"{años} años", ALx/años, des/años, mod/años, costo
            ]
df_costo=pd.DataFrame(data=respuestas, columns=columnas, index=filas)
#Mostrar dataframe
print("El costo de x a x+1:")
df_costo
```

```python
#Calculo de datos para la situación esperada en x+1
AL_x_plus_1 = (ALx + NCx) * (1 + inte)
F_x_plus_1 = costo + Fx * (1 + inte)
UAL_x_plus_1_e = F_x_plus_1 - AL_x_plus_1
UAL_cero_x_plus_1 = ALx - ALx / años
#Definir dataframe
filas = ['AL_(x+1)', 'F_(x+1)', 'UAL_{x+1}^{e}', 'UAL_{cero}^{x+1}']
columnas = ['Respuesta']
#Asignar valores calculados
respuestas = [AL_x_plus_1,
              F_x_plus_1,
              UAL_x_plus_1_e,
              UAL_cero_x_plus_1
              ]
df_esperado = pd.DataFrame(data=respuestas, columns=columnas, index=filas)
#Mostrar dataframe
print("Situacion esperada en x+1:")
df_esperado
```
